```python
def authenticate_youtube():
    """
    Authenticate with YouTube API using OAuth 2.0
    Returns: YouTube API credentials
    """
    # The token file stores the user's access and refresh tokens
    token_file = 'token.pickle'
    credentials = None

    # Scopes required for the YouTube API
    SCOPES = [
        'https://www.googleapis.com/auth/youtube.force-ssl',
        'https://www.googleapis.com/auth/youtube'
    ]

    # Load existing credentials from token.pickle if it exists
    if os.path.exists(token_file):
        print('Loading existing credentials...')
        with open(token_file, 'rb') as token:
            credentials = pickle.load(token)

    # If there are no valid credentials available, prompt the user to log in
    if not credentials or not credentials.valid:
        if credentials and credentials.expired and credentials.refresh_token:
            print('Refreshing access token...')
            try:
                credentials.refresh(Request())
            except Exception as e:
                print(f'Error refreshing token: {e}')
                credentials = None
        
        if not credentials:
            print('Fetching new tokens...')
            try:
                # Load client secrets from your client_secrets.json file
                flow = InstalledAppFlow.from_client_secrets_file(
                    'client_shh.json', SCOPES)
                
                # Run local server to receive authentication code
                credentials = flow.run_local_server(port=0)

                # Save the credentials for future use
                with open(token_file, 'wb') as token:
                    pickle.dump(credentials, token)
                print('New credentials saved successfully!')
            
            except Exception as e:
                print(f'Error during authentication: {e}')
                raise

    return credentials
```

# Get comment thread 

```python
def parse_comments(response):
    # Prase comments
    comments = []
    for comment in response["items"]:

        comment_id = comment["snippet"]["topLevelComment"]["id"]
        top_comment = comment["snippet"]["topLevelComment"]["snippet"]    
        comments.append(
            {
                "id" : comment_id,
                "parent_comment_id" : None,
                "author" : top_comment["authorDisplayName"],
                "author_channel_id" : top_comment["authorChannelId"]["value"],
                "text" : top_comment["textDisplay"]
            }
        )
    return comments
```

```python
def get_comment_thread(video_id:str, is_filter=True, max_results=20, pages=1):
    # Check if trying to pull bulk data without filtering
    if pages > 2 and is_filter == False:
        print ("Error: pages > 2 and is_filter=False. Use filter for large queries.")
        return None
    
    args = {
        "part" : "snippet",
        "maxResults" : max_results,
        "videoId" : video_id
    }
    if is_filter:
        #args["fields"] = "items/id,snippet/topLevelComment/snippet/textDisplay,authorDisplayName,likeCount"
        comment_cols = "textDisplay,authorChannelId,authorDisplayName,likeCount"
        args["fields"] = f"(nextPageToken,items(snippet/topLevelComment(id,snippet({comment_cols}))))"
    
    request = youtube.commentThreads().list(**args)
    response = request.execute()
    
    #return response
    
    comments = parse_comments(response)
    
    if pages > 1:
        for index in range(pages):
            # Extract next page token
            next_page_token = response["nextPageToken"]
            args["pageToken"] = next_page_token
            request = youtube.commentThreads().list(**args)
            response = request.execute()
            comments.extend(parse_comments(response))
    
    comments = pd.DataFrame(comments)
    
    
    return comments
```
