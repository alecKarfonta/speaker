# TTS API

```python
host = "192.168.1.196"
port = 8401
endpoint = f"http://{host}:{port}/"
response = requests.get(endpoint)
response.status_code
```

```python
# list voices
endpoint = f"http://{host}:{port}/voices"
response = requests.get(endpoint)
response.status_code
response.json()
```

```python
def tts_api(text, voice_name, language, tau=0.6, gpt_cond_len=10, top_k=10, top_p=0.9, decoder_iterations=10, split_sentences=True):
    endpoint = f"http://{host}:{port}/tts"
    payload = {
        "text" : text,
        "voice_name" : voice_name
    }
    response = requests.post(endpoint, json=payload)
    response.status_code

    wav = np.frombuffer(response.content, dtype=np.float32)
    return wav
```

```python
text = "this is a test" 

wav = tts_api(text, "biden", "en")
IPython.display.Audio(wav, rate=24000)
```

```python
text = "Quiero decir, cuando se considera que utiliza Zen 5 frente a la Zen 3 vamos a utilizar en este sistema. Uno pensaría que la respuesta sería obvia." 

wav = tts_api(text, "batman", "es")
IPython.display.Audio(wav, rate=24000)
```

```python
text = "Speaking of motherboard, this ASUS Prime Be 550 M A is a pretty good value, when it's on sale for around $100."
voice = "trump_01"
wav = tts_api(text, voice, "en", tau=0.3, gpt_cond_len=2, top_k=10, top_p=0.9, decoder_iterations=25, split_sentences=True)

# Save the audio to file
sf.write(f'notebooks/test_{voice}.wav', wav, 24000)

IPython.display.Audio(wav, rate=24000)
```

```python
text = "Quer construir um PC de jogos doentio com um orçamento razoável, mas a questão é que mil dólares não se estendem tanto como costumavam."

wav = tts_api(text, "trump", "pt")
IPython.display.Audio(wav, rate=24000)
```
