```python
import numpy as np
import requests
import soundfile as sf
```

```python
def tts_api(text, voice_name, language):
    endpoint = f"http://{host}:{port}/tts"
    payload = {
        "text" : text,
        "voice_name" : voice_name,
        "language" : language
    }
    response = requests.post(endpoint, json=payload)
    response.status_code

    wav = np.frombuffer(response.content, dtype=np.float32)
    return wav
```

```python
text = "You want to build a sick gaming PC on a reasonable budget, but the thing is, a thousand bucks doesn't stretch as far as it used to." 
host = "192.168.1.196"
port = 8001
voice = "biden"
endpoint = f"http://{host}:{port}/tts"
payload = {
    "text" : text,
    "voice_name" : voice
}
response = requests.post(endpoint, json=payload)
response.status_code

wav = np.frombuffer(response.content, dtype=np.float32)
# Write audio to file
sf.write(f'notebooks/test_{voice}.wav', wav, 24000)
IPython.display.Audio(wav, rate=24000)
```

```python
def tts(text):
    host = "threadripper"
    port = 8001
    endpoint = f"http://{host}:{port}/tts"
    payload = {
        "text" : text
    }
    response = requests.post(endpoint, json=payload)
    response.status_code

    wav = response.json()["wav"]
    samplerate = response.json()["samplerate"]
    return wav, samplerate

```

```python
text = """I guess Biden's new slogan is 'Sleepy Joe' for a reason!" 😂 #SelfCarePresident"""
wav, samplerate = tts(text)
IPython.display.Audio(wav, rate=samplerate/2, autoplay=True)
```

```python
def chatbot_tts(text, person="Alec"):
    host = "threadripper"
    # Hit chatbot
    port = 5001
    payload = {
        "comment": text,
        "commentor": person
    }
    endpoint = f"http://{host}:{port}/comment"
    response = requests.post(endpoint, json=payload)
    #print(response.json()["response"])

    text = response.json()["response"]
    #text = "Let's make that our password, for when we meet again." # And where shall I go now? The net is vast and limitless."
    #print (text)  
    # Hit tts
    port = 8001
    endpoint = f"http://{host}:{port}/tts"
    payload = {
        "text" : text
    }
    response = requests.post(endpoint, json=payload)
    response.status_code

    wav = response.json()["wav"]
    samplerate = response.json()["samplerate"]

    return text, wav, samplerate
```
